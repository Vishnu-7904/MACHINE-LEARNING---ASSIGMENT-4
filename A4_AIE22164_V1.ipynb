```python
# Importing all the  required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.impute import SimpleImputer
from scipy.spatial import distance



# QUESTION 1

df = pd.read_excel("C:/Users/91914/Downloads/weatherAUS.xlsx") # Read the data from the specified Excel file into a DataFrame.
def mean(Class): # I am taking WindGustSpeed and WindSpeed9am
    mean_class = df[Class].mean(axis=0) #Calculating the mean
    return mean_class

def standard_deviation(Class):
    std_deviate = df[Class].std(axis=0) #Calculating the standard deviation
    return std_deviate

def Distance():
    df1 = pd.DataFrame(df)
    Array = df1.to_numpy() # Converting dataframe to array
    distance = np.linalg.norm(mean('WindSpeed9am') - mean('WindGustSpeed'))
    return distance

# Printing all the mean and standard deviation
print("Mean of the class WindSpeed9am:", mean('WindSpeed9am'))
print("Mean of the class WindGustSpeed:", mean('WindGustSpeed'))
print("Standard deviation of WindSpeed9am:", standard_deviation('WindSpeed9am'))
print("Standard deviation of WindGustSpeed:", standard_deviation('WindGustSpeed'))
print("Distance between mean vectors:", Distance())



# QUESTION 2

mean_WindGustSpeed = df['WindGustSpeed'].mean() # Calculating mean and variance
variance_WindGustSpeed = df['WindGustSpeed'].var()
mpl.hist(df['WindGustSpeed'], bins=20, color='blue', alpha=0.7) # Ploting the histogram
mpl.xlabel('WindGustSpeed')
mpl.ylabel('Frequency')
mpl.title('Histogram of WindGustSpeed')
mpl.show()

# Printing the Mean and Variance
print("Mean WindGustSpeed:", mean_WindGustSpeed)
print("Variance WindGustSpeed:", variance_WindGustSpeed)



#QUESTION 3 and QUESTION 4

X = df[['WindGustSpeed', 'WindSpeed9am']] # Merge features into one array
imp = SimpleImputer(strategy="mean") # Impute missing values with mean for each column
X_imputed = imp.fit_transform(X)
y = df['RainTomorrow'] # Impute missing values with most frequent label
imp2 = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
y_imputed = imp2.fit_transform(y.values.reshape(-1, 1)).flatten()

minkowski_distance = distance.minkowski(X_imputed[0], X_imputed[1]) # Calculating Minkowski distance
print("Minkowski distance:", minkowski_distance)
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.5) # Spliting the data into train and test sets

X = df[['WindGustSpeed', 'WindSpeed9am']] # Merging both the features into one array

imp = SimpleImputer(strategy="mean") # Impute missing values with mean
X_imputed = imp.fit_transform(X)
y = df['RainTomorrow'] # Impute missing values with most frequent label
imp2 = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
y_imputed = imp2.fit_transform(y.values.reshape(-1, 1)).flatten()



# QUESTION 5

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.5) # Spliting the data into train and test sets
neigh = KNeighborsClassifier(n_neighbors=3) # Applying KNN of K=3
neigh.fit(X_train, y_train) # Training the model



# QUESTION 6

accuracy = neigh.score(X_test, y_test) # Returning the accuracy of the model
print("Accuracy:", accuracy)


# QUESTION 7

test_vector = neigh.predict(X_test) # Returning the prediction vector predicted by the model
print("Test vector (sample):", test_vector[0:100])


# QUESTION 8

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.5) # Comparing the accuracy of KNN with K=1 and K=3
neigh_1 = KNeighborsClassifier(n_neighbors=1) # Create a KNN classifier object with K=1 neighbors
neigh_1.fit(X_train, y_train) # Train the KNN classifier using the training data
accuracy_1 = neigh_1.score(X_test, y_test) # Calculating the accuracy for K=1
print("K=1 accuracy:", accuracy_1)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.5) # Split the data into train and test sets for K=3
neigh_3 = KNeighborsClassifier(n_neighbors=3) # Create a KNN classifier object with K=3 neighbors
neigh_3.fit(X_train, y_train) # Train the KNN classifier using the training data
accuracy_3 = neigh_3.score(X_test, y_test) # Calculating the accuracy for K=3
print("K=3 accuracy:", accuracy_3)
accuracy_list = []

# Finding the accuracy from K=1 to K=11
for k in range(1, 12):
    X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.5)
    neigh = KNeighborsClassifier(n_neighbors=k)  # Create a KNN classifier object with current value of K
    neigh.fit(X_train, y_train)  # Train the KNN classifier using the training data
    accuracy = neigh.score(X_test, y_test)  # Calculating accuracy for current value of K
    accuracy_list.append(accuracy)

# Plotting the accuracy
mpl.plot(range(1, 12), accuracy_list)
mpl.xlabel('k')
mpl.ylabel('Accuracy')
mpl.title("Accuracy plot for different values of k")
mpl.show()


# QUESTION 9

# Returning the confusion matrix and the classification report
# Printing the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, test_vector))
# Printing the classification report
print("\nClassification Report:")
print(classification_report(y_test, test_vector))
```
